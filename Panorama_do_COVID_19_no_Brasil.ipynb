# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://raw.githubusercontent.com/carlosfab/dsnp2/master/img/covid19.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados para o desenvolvimento desse projeto foram disponibilizados pela organização [*Our World in Data*](https://ourworldindata.org/), uma entidade digital **especializada** em divulgar pesquisas referentes à diversos temas em **nível mundial**. Mais precisamente é possível encontrar a planilha através deste [repositório](https://github.com/owid/covid-19-data/tree/master/public/data) onde se encontram consolidadas todas as informações referentes ao *dataset* disponibilizado pela organização. 

Trata-se uma amostra complexa de dados devido à variedade de fontes de onde as informações foram coletadas.

Os dados disponibilizados foram extraídos de diversas fontes privadas, governamentais e não governamentais. Exemplo:

* **Casos confirmados de COVID-19 e mortes**: Centro de Ciência e Engenharia de Sistemas (CCES) da **Universidade Jhons Hopkins**

* **Hospitalizações e admissões em Unidades de Terapia Intensivas**: Fontes oficiais e pesquisas da organização *Our World in Data*.
* **Testes de COVID-19**: Fontes oficiais e pesquisas da organização *Our World in Data*.
* **Vacinações contra COVID-19**: Fontes oficiais e pesquisas da organização *Our World in Data*. 
* **Outras variáveis**: Dados coletados de várias fontes (Nações Unidas, Banco Mundial, Carga Global de Morbidade (GBD)).

  * Total de mortes confirmadas por um milhão de pessoas.
* **new_deaths_per_million**

* **new_deaths_smoothed_per_million**

* **reproduction_rate**

* **hosp_patients_per_million**
  * Número de pacientes com COVID-19 por um milhão em hospitais.
* **weekly_icu_admissions**
  * Número de pacientes que deram entrada em Unidades de Terapia Intensivas em uma semana (na data atual e nos últimos seis dias).
* **weekly_icu_admissions_per_million**
  * Número de pacientes por um milhão que deram entrada em Unidades de Terapia Intensivas em uma semana (na data atual e nos últimos seis dias).
* **weekly_hosp_admissions**

  * Número de pacientes por um milhão que deram entrada em hospitais em uma semana (na data atual e nos últimos seis dias).

* **total_tests_per_thousand**

* **new_tests_per_thousand**

* **tests_units**

* **total_vaccinations**

* **people_vaccinated**

  * Total de doses de reforço administradas.
* **new_vaccinations**

  * Total de doses de vacinas administradas por cem pessoas.

  * Total de pessoas vacinadas que receberam ao menos uma dose  por cem pessoas.
* **people_fully_vaccinated_per_hundred**
  * Total de pessoas vacinadas que receberam todas as doses por cem pessoas.
* **total_boosters_per_hundred**
  * Total de doses de reforço da vacina administrada por cem pessoas.
* **new_vaccinations_smoothed_per_million**
  * Novas doses de vacina administradas por um milhão do total da população.
* **new_people_vaccinated_smoothed**
  * Total diário de pessoas que receberam a primeira dose nos últimos sete dias.
* **new_people_vaccinated_smoothed_per_hundred**
  * Total diário de pessoas que receberam a primeira dose nos últimos sete dias por cem pessoas do total da população.
* **stringency_index**
  * Índice de Rigidez de resposta do governo. Refere-se às medidas governamentais para contenção do COVID-19.
* **population**
  * População total.
* **population_density**
  * Número de pessoas divididas por área, medido em quilômetros quadrados.
* **median_age**
  * Média de idade da população
* **aged_65_older**
  * Parte da população acima de 65 anos.
* **aged_70_older**
  * Parte da população acima de 70 anos.
* **gdp_per_capita**
  * Produto Interno Bruto (PIB) per capita (por pessoa).

  * Parte da população vivendo em pobreza extrema
* **cardiovasc_death_rate**[
  * Taxa de morte por complicações cardiovasculares em 2017.
* **diabetes_prevalence**
  * Prevalência de diabetes em 2017.
* **female_smokers**
  * Parte da população feminina fumante (dados do último ano disponível).
* **male_smokers**
  * Parte da população masculina fumante (dados do último ano disponível).
* **handwashing_facilities**
  * Parte da população com higiene básica.
* **hospital_beds_per_thousand**
  * Número de leitos de hospitais por um milhão de pessoas.
* **life_expectancy**
  * Expectativa de vida de nascimento em 2019.
* **human_development_index**
  * Índice de desenvolvimento humano, baseado em expectativa de vida, educação e renda.
* **excess_mortality**
  * Diferença percentual entre o número de mortes semanais ou mensais em 2020-2021  e o número projetado de mortes para o mesmo período baseada em anos anteriores.
* **excess_mortality_cumulative**
  * Diferença percentual entre o número de mortes acumuladas de 1º de janeiro de 2020 e as mortes cumulativas projetadas para o mesmo período baseada em anos anteriores.
* **excess_mortality_cumulative_absolute**
   * Diferença cumulativa entre o número relatado de óbitos desde 1º de janeiro de 2020 e o número projetado de óbitos para o mesmo período com base em anos anteriores
* **excess_mortality_cumulative_per_million**
  * Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base em anos anteriores, por milhão de pessoas


### Importando os dados do COVID-19

```
# importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# importar o dataset COVID-19 para um DataaFrame

FILE_PATH = (
    "https://raw.githubusercontent.com/owid/covid-19-data"
    "/master/public/data/owid-covid-data.csv"
)

df = pd.read_csv(FILE_PATH)
```

### Análise Exploratória para o Brasil

Nesta seção restringiremos a análise aos dados referentes ao cenário **brasileiro**.

```
df_brasil = df.loc[df["iso_code"] == "BRA"].copy()
df_brasil.head()
```

```
# dimensoes do dataframe
rows, cols = df_brasil.shape
old_qtd_cols = cols
print("Linhas:", rows)
print("Colunas:", cols)
```

```
df_brasil.info()
```

```
# como observado em analise, ha colunas sem dados, portanto serao removidas
nan_values = float("NaN")
df_brasil.replace("", nan_values, inplace=True)
df_brasil.dropna(how='all', axis=1, inplace=True)

new_rows, new_cols = df_brasil.shape
print(f"Quantidade de colunas removidas: {old_qtd_cols - new_cols}")
```

> Como foi observado através do método head(), o *dataset* possui valores `NaN`, que deixam a visualização **poluída**. Tendo em vista a natureza das variáveis, é possível substituir esses valores por `0`, a fim de tornar mais limpa nossa base de dados.

```
df_brasil.fillna(0, axis=1, inplace=True)
```

```
# Nova visualizacao apos remocao das colunas
df_brasil.head()
```

## Análise exploratória estatística

Nesta seção iremos iniciar a análise estatística dos dados, buscando **extrair** medidas descritivas, comparações entre as variáveis que compõem o *dataset* e porventura alguns dados interessantes de serem observados. 


```
#primeiros deixaremos separado as colunas que serão interessantes para analisar separadamente
selected_columns = ["total_cases", "new_cases", "total_cases_per_million", "total_deaths",
                    "new_deaths", "new_deaths_per_million","reproduction_rate","total_tests",
                    "new_tests","total_vaccinations", "people_vaccinated", 
                    "people_fully_vaccinated", "total_boosters", "new_vaccinations", "stringency_index"]
```

No seguinte gráfico de barras, visualizaremos a quantidade total de pessoas que tomaram ao menos uma dose, o total de casos de COVID-19 registrados, o total de mortes, o total de doses de reforço administradas e a quantidade de testes realizados.

### Primeira morte registrada

### Primeira pessoa vacinada

```
first_vaccination = (df_brasil.loc[df_brasil.new_vaccinations >= 1.0]).sort_values(by="new_vaccinations", ascending=True).index[0]
first_vaccination.date().strftime('%Y-%m-%d')
```

### Tempo entre o registro do primeiro caso e a primeira morte

> Antes de partimos para as seguintes **análises gráficas** comparativa entre as variáveis, primeiramente temos que **normalizar os dados**, pois algumas varáveis possuem escalas diferentes.

```
#normalizar os dados
for column in df_brasil[selected_columns].columns:
    df_brasil[column] = df_brasil[column]  / df_brasil[column].abs().max()
```

> Um dado interessante de ser observado é a **frequência médias** das variáveis a serem analisadas. Analisaremos `total_cases`, `new_cases`, `total_cases_per_million`,`total_deaths`, `new_daths`, `new_deaths_per_million`, `reproduction_rate`, `total_tests`, `new_tests`, `total_vaccinations`, `people_vaccinated`, `people_fully_vaccinated`, `total_boosters`, `new_vaccinations`, `stringency_index`, conforme **separadas anteriormente**. Pelos **histogramas** que serão plotados é possível, como mencionado, observar a frequência de distribuição e eventualmente a presença de alguns *outliers*.


```
df_frequency = df_brasil[selected_columns]
                                
df_frequency.hist(bins=15, figsize=(15,10))
plt.show()
```

> Outra medida descritiva interessante de se observar é a **correlação entre as variáveis**, ainda mais nesse cenário onde podemos notar uma nítida ligação entre os casos de morte, quantidade de vacinados, surgimento de novos casos de COVID-19 entre outros. Restringiremos a análise há algumas variáveis, sendo elas as mesmas observadas nos histogramas.

```
#criando correlacao
df_brasil_corr = df_brasil[selected_columns].corr()

fig, ax = plt.subplots(figsize=(11,11))  

sns.heatmap(df_brasil_corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True, ax=ax);
```

### Correlações entre as variáveis

> Agora visualizaremos as medidas descritivas tabulares e gráficas para as variáveis `total_vaccinations`, `total_deaths`, `total_cases` que possuem grande relação entre si.

```
df_brasil[["total_vaccinations", "total_deaths", "total_cases"]].describe()
```

```
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(15,5))  

sns.boxplot(data=df_brasil.loc['2021-01-01':'2021-12-31', "total_cases"], 
            ax=ax3, 
            width=0.4,
            color="darkred")

sns.set_palette("pastel")            
sns.boxplot(data=df_brasil.loc['2021-01-01':'2021-12-31', "total_deaths"], 
            ax=ax2,
            width=0.4,
            color="green")

sns.boxplot(data=df_brasil.loc['2021-01-01':'2021-12-31', "total_vaccinations"], 
            ax=ax1, 
            width=0.4,
            color="skyblue")

ax1.set_title("Vacinações")
ax2.set_title("Mortes")
ax3.set_title("Casos de COVID-19")

ax1.tick_params(
    axis='x',           
    bottom=False,      
    labelbottom=False) 
ax2.tick_params(
    axis='x',           
    bottom=False,      
    labelbottom=False) 
ax3.tick_params(
    axis='x',           
    bottom=False,      
    labelbottom=False) 

plt.show()
```

> O período mais **polêmico** da pandemia em território nacional certamente foi o ano de 2021. Foi em tal ano que vivemos o auge da pandemia e suas crises. No gráfico a seguir buscaremos analisar como o governo da federação reagiu à situação no intervalo de 1 ano e meio (de janeiro de 2021 - junho de 2022).

>**PS**: A variável `stringency_index`, conforme exposto no dicionário de variáveis, refere-se à avaliação de como o *governo* respondeu à pandemia, sendo avaliado de **0 a 100** sendo 100 a resposta mais rigorosa.

```
fig, ax = plt.subplots(figsize=(10,5))  
(df_brasil.loc['2021-01-01':'2022-06-30', "stringency_index"]).plot(kind="line", ax=ax, label="Avaliação das medidas governamentais")
(df_brasil.loc['2021-01-01':'2022-06-30', "new_cases"]).plot(kind="line", ax=ax, label="Novos casos")

plt.title("Avaliação das medidas governamentais X Novos casos de COVID-19")
plt.legend(loc='best')
plt.tight_layout()
```

No seguinte gráfico será representado a evolução dos indicadores de total de morte, total de casos e pessoas completamente vacinadas. Será possível notar a **correlação** entre a visualização **anterior e a seguinte**

```
# Evolução dos totais de casos, mortes e vacinação completa durante o ano de 2021

fig, ax = plt.subplots(figsize=(15,5))  
df_brasil.groupby(pd.Grouper(freq='M', offset=5, label='right')).first()

(df_brasil["total_deaths"]).plot(kind="line", ax=ax, label="Total de mortes")
(df_brasil["total_cases"]).plot(kind="line", ax=ax, label="Total de casos")
(df_brasil.loc[df_brasil.people_fully_vaccinated > 0.00]).people_fully_vaccinated.plot(kind="line", ax=ax, label="Pessoas completamente vacinadas")

plt.legend(loc='best') 
plt.xlabel("Data")
plt.ylabel("Total")
plt.title("Total de mortes X Total de casos X Pessoas completamente vacinadas")
plt.tight_layout()
```

> Nos seguintes *violin plots* será exibido algumas das **medidas descritivas** (mediana, assimetrias, mínimo, máximo) das variáveis em questão a fim de enriquecer a visualização e noção de proporcionalidade (correlação) das mesmas.

```
fix, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12,4), constrained_layout=True)

sns.violinplot(data=df_brasil, x="total_tests", ax=ax1, color="skyblue", cut=0)
sns.violinplot(data=df_brasil, x="new_vaccinations", ax=ax2)
sns.violinplot(data=df_brasil, x="new_cases", ax=ax3, color="red")

ax1.set_title('Total de testes')
ax2.set_title('Novas vacinações')
ax3.set_title('Novos casos')

ax1.set_xlabel("u.m")
ax2.set_xlabel("u.m")
ax3.set_xlabel("u.m")

plt.show()
```

```
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,4), constrained_layout=True)

df_people_vaccinated = df_brasil.loc['2021-03-01':'2022-03-01', 'people_fully_vaccinated']
df_total_deaths = df_brasil.loc['2021-03-01':'2022-03-01', 'total_deaths']
sns.violinplot(x=df_people_vaccinated, ax=ax1, color="green")
sns.violinplot(x=df_total_deaths, ax=ax2, color="gray")

fig.suptitle('Distribuição estimativa', fontsize=18)
ax1.set_title('Pessoas completamente vacinadas')
ax2.set_title('Total de mortes')
ax1.set_xlabel("u.m")
ax2.set_xlabel("u.m")
plt.show()
```
